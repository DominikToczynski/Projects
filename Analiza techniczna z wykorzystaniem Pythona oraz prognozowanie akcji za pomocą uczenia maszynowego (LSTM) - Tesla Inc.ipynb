## 1. Analiza technicza

Analiza techniczna to jedna z dyscyplin handlowych, która wykorzystywana jest do oceny inwestycji oraz identyfikacji możliwości
handlowych na podstawie ruchu ceny, kształtu wykresu oraz wolumenu wybranego instrumentu handlowego. Analiza techniczna opiera
sie na danych historycznych starając się, poprzez zastosowanie powyższych metod, przewidzieć ruch kursu analizowanego 
instrumentu.

Analiza techniczna jest często używana do generowania krótkoterminowych sygnałów transakcyjnych korzystając z różnych narzędzi 
do tworzenia wykresów, ale może również pomóc w poprawie oceny siły lub słabości papieru wartościowego w stosunku do szerszego 
rynku lub jednego z jego sektorów. Informacje te pomagaja analitykom poprawić ogólną ocenę wyceny analizowanego instrumentu.

Profesjonalni analitycy często wykorzystują analizę techniczną w połączeniu z innymi formami badań. Indywidualni inwestorzy
mogą podejmować decyzje wyłącznie na podstawie wykresów cen papierów wartościowych i podobnych statystyk, ale praktykujący 
analitycy giełdowi rzadko ograniczają swoje badania wyłącznie do analizy fundamentalnej lub technicznej.

Analizę techniczną można zastosować do dowolnego papieru wartościowego z historycznymi danymi handlowymi. Obejmuje to akcje, 
kontrakty terminowe, towary, stały dochód - instrumenty dłużne, waluty i inne papiery wartościowe. W rzeczywistości analiza 
techniczna jest znacznie bardziej rozpowszechniona na rynkach towarowych i forex, na których inwestorzy koncentrują się na 
krótkoterminowych ruchach cen.

Analiza techniczna najczęściej dotyczy zmian cen, ale niektórzy analitycy śledzą liczby inne niż tylko ceny, takie jak wolumen 
obrotu lub liczby otwartych pozycji.

## 1.2 Ograniczenia analizy technicznej

Niektórzy analitycy i badacze akademiccy spodziewają się, że hipoteza efektywnego rynku pokazuje, dlaczego nie powinno się
oczekiwać, że w historycznych danych dotyczących cen i wolumenu będą zawarte jakiekolwiek przydatne informacje. Te punkty 
widzenia są znane jako słaba forma i pół-silna forma hipotezy efektywnego rynku.

Inną krytyką analizy technicznej jest to, że historia nie powtarza się dokładnie, więc badanie wzorców cen ma wątpliwe znaczenie
i można je zignorować. Ceny wydają się lepiej modelowane przy założeniu losowego spaceru - odrzucenie analizy technicznej i 
fundamentalnej, na rzecz długoterminowych inwestycji, które odwzorowują działania indeksów giełdowych (paswyne zarządzanie).

Trzecią krytyką analizy technicznej jest to, że w niektórych przypadkach działa, ale tylko dlatego, że stanowi samospełniającą 
się przepowiednię. Na przykład sytuacja, w której wielu technicznych traderów złoży zlecenie stop-loss poniżej 200-dniowej średniej kroczącej określonej firmy. Jeśli zrobiła to duża liczba traderów, a akcje osiągną tę cenę, pojawi się duża liczba zleceń sprzedaży, które popchną akcje w dół, potwierdzając ruch oczekiwany przez traderów. Wtedy inni handlowcy zobaczą spadek ceny, więc także sprzedają swoje pozycje, wzmacniając siłę trendu. Tę krótkoterminową presję na sprzedaż można uznać za samospełniającą się, ale będzie ona miała niewielki wpływ na to, jaka cena aktywów nastąpi za kilka tygodni lub miesięcy. Podsumowując, jeśli wystarczająca liczba osób użyje tych samych sygnałów, mogą spowodować ruch przepowiedziany przez sygnał.

## 3.7. MACD (Moving Average Convergence Divergence)

MACD to wskaźnik momentum podążający za trendem, który pokazuje związek między dwiema średnimi ruchomymi ceny papieru 
wartościowego. MACD jest obliczany przez odjęcie 26-okresowej wykładniczej średniej kroczącej (EMA) od 12-okresowej EMA.

Wynikiem tych obliczeń jest linia MACD. Dziewięciodniowa EMA MACD, zwana „linią sygnału”, jest następnie wykreślana na górze 
linii MACD, która może działać jako wyzwalacz sygnałów kupna i sprzedaży. Handlowcy mogą kupować papier wartościowy, gdy MACD 
przecina się powyżej linii sygnałowej i sprzedawać (lub grać na sprzedaż - short sell) papier wartościowy, gdy MACD przecina 
się poniżej linii sygnałowej. Wskaźniki rozbieżności średniej ruchomej zbieżności (MACD) można interpretować na kilka sposobów, 
ale bardziej powszechnymi metodami są skrzyżowania, rozbieżności i szybkie wzrosty / spadki.

Formuła dla MACD to:
    
MACD = 12-Period EMA − 26-Period EMA

MACD oblicza się, odejmując długoterminową EMA (26 okresów) od krótkoterminowej EMA (12 okresów). Wykładnicza średnia krocząca 
(EMA) to rodzaj średniej ruchomej (MA), która przypisuje większą wagę i znaczenie najnowszym punktom danych.

--------------------------------------------------------------------------------------------------------------------------------

https://economictimes.indiatimes.com/definition/moving-average-convergence-divergence

https://www.investopedia.com/terms/m/macd.asp

```python

exp1 = tsla['Close'].ewm(span=12, adjust=False).mean()
exp2 = tsla['Close'].ewm(span=26, adjust=False).mean()
tsla['MACD'] = exp1 - exp2
tsla['Signal Line'] = tsla['MACD'].ewm(span=9, adjust=False).mean()
```

```python
fig, ax = plt.subplots(figsize=(15,7))
tsla[['MACD', 'Signal Line']].plot(ax=ax)
tsla['Close'].plot(ax=ax, alpha=0.25, secondary_y=True)
```

Sposób, w który należy interpretować powyższy wykres jest następujący. W momencie, w którym niebieska linia MACD znajduje się 
nad pomarańczową linią Signal Line, jest to sygnał do tego, aby kupić akcje lub zatrzymać te, które aktualnie posiadamy. 
Natomiast w sytuacji odwrotnej, kiedy linia Signal Line znajduje się nad MACD jest do sygnał do tego, aby akcje sprzedać.

Kiedy porówany to co dzieje się na wykresie pomiedzy obiema liniami, a tym co dzieje sie na zielonej lini, która reprezentuje
ceny zamknięcia akcji Tesli w okresie 2021 roku, będziemy mogli zaobserować, że sygnały, które dostajemy od MACD oraz Signal Line, dobrze interpretują rzeczywisty ruch ceny akcji.

## 3.8 Oscylator stochastyczny

Oscylator stochastyczny to wskaźnik momentum porównujący określoną cenę zamknięcia papieru wartościowego z zakresem jego cen 
w określonym przedziale czasu. Wrażliwość oscylatora na ruchy rynkowe można zredukować, dostosowując dany okres lub biorąc 
średnią ruchomą wyniku. Służy do generowania sygnałów transakcyjnych dotyczących wykupienia i wyprzedania, wykorzystując 
ograniczony zakres wartości od 0 do 100.


Formuła dla oscylatora stochastycznego to:
    
%K = (C - L14 / H14 - L14) * 100

Gdzie:
    
C - najnowsza cena zamknięcia

L14 - najniższa cena w obrocie z 14 poprzednich sesji handlowych

H14 - najwyższa cena handlowana w tym samym czasie za okres 14 dni

%K - aktualna wartość wskaźnika stochastycznego


Ogólna teoria służąca jako podstawa dla tego wskaźnika jest taka, że na rynku z tendencją wzrostową ceny zamkną się blisko 
szczytu, a na rynku z tendencją spadkową ceny blisko dołka. Sygnały transakcyjne są tworzone, gdy %K przecina trzyokresową 
średnią kroczącą, zwaną %D

Oscylator stochastyczny jest ograniczony zakresem, co oznacza, że zawsze wynosi od 0 do 100. To czyni go użytecznym 
wskaźnikiem warunków wykupienia i wyprzedania. Tradycyjnie odczyty powyżej 80 są uważane za wykupione, a odczyty 
poniżej 20 są uważane za wyprzedane. Jednak nie zawsze wskazują one na zbliżające się odwrócenie; bardzo silne trendy mogą 
utrzymywać warunki wykupienia lub wyprzedania przez dłuższy czas. Zamiast tego inwestorzy powinni zwracać uwagę na zmiany w 
oscylatorze stochastycznym w poszukiwaniu wskazówek dotyczących przyszłych zmian trendów.

Stochastyczny wykres oscylatora składa się zazwyczaj z dwóch linii: jednej odzwierciedlającej rzeczywistą wartość oscylatora 
dla każdej sesji, a drugiej odzwierciedlającej jego prostą średnią kroczącą z trzech dni. Ponieważ uważa się, że cena 
podąża za pędem, przecięcie tych dwóch linii jest uważane za sygnał, że może trwać odwrócenie, ponieważ wskazuje na dużą
zmianę pędu z dnia na dzień.

--------------------------------------------------------------------------------------------------------------------------------

https://economictimes.indiatimes.com/definition/stochastic-oscillator

https://www.investopedia.com/terms/m/macd.asp

```python
high14 = tsla['High'].rolling(14).max()
low14 = tsla['Low'].rolling(14).min()
tsla['%K'] = (tsla['Close'] - low14)*100/(high14-low14)
tsla['%D'] = tsla['%K'].rolling(3).mean()
```

```python
fix, ax = plt.subplots(figsize=(18,9))
tsla[['%K', '%D']].plot(ax=ax)
tsla['Close'].plot(ax=ax, alpha=0.25, secondary_y=True)
ax.axhline(80, c='r', alpha=0.3)
ax.axhline(20, c='r', alpha=0.3)
```

Sposób, w który należy interpretować powyższy wykres jest następujący. Kiedy linia %D przecina się i przesuwa się poniżej lini 
%K można to odczytać jako sygnał do sprzedaży, szczególnie jeżeli taka sytuacja zachodzi powyżej lini poziomej czerownej przy wartościach 80+ - wtedy można powiedzieć, że instrument jest 'wykupiony'. O wykupieniu można mówić, gdy uważa się, że papier wartościowy jest sprzedawany na poziomie powyżej jego wartości godziwej. Wykupienie generalnie opisuje niedawne lub krótkoterminowe zmiany ceny papieru wartościowego i odzwierciedla oczekiwanie, że rynek poprawi cenę w najbliższej przyszłości. Natomiast kiedy linia K% przecina się i przesuwa powyżej lini %D, można to interpretować jako sytuacje, w której kurs rynkowy
rośnie szybciej niż średia reprezentowana przez linię %D, ten wzrot jest uważany za sygnał kupna.

Kiedy porówany to co dzieje się na wykresie pomiedzy obiema liniami, a tym co dzieje sie na zielonej lini, która reprezentuje
ceny zamknięcia akcji Tesli w okresie 2021 roku, będziemy mogli zaobserować, że sygnały, które dostajemy od ruchu lini %K 
oraz %D, dobrze interpretuja rzeczywisty ruch ceny akcji.

```python

training_set = tsla_lstm['Close']
training_set = pd.DataFrame(training_set)
```

## 5.2. Wstępne przetwarzanie danych

```python
# Sprawdzamy czy w naszym zbiorze znajdują się jakieś brakujące wartości

tsla_lstm.isna().any()
```

Feature Scaling. Skalowanie danych do zakresu od 0 do 1 powinno przyczynić się do tego, że otrzymamy mniejsze odchylenia standardowe, które mogą tłumić efekt wartości odstających.

```python

from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler(feature_range = (0,1))
training_set_scaled = sc.fit_transform(training_set)
```

W tym miejscu przygotujemy nasz zestaw x_train oraz y_train. Dane będą przechodzić od 1 do 60 dnia i tworzyć predykcje na dzień 61, następnie przechodzić od 2 do 62 dnia i tworzyć predykcje na dzień 62. Dane zostaną dodane do obu list, a następnie każda z nich zostanie przekształcona

```python
# Przygotowanie x_train oraz y_train

x_train = []
y_train = []

for i in range(60, 1258):
    x_train.append(training_set_scaled[i-60:i, 0])
    y_train.append(training_set_scaled[i, 0])
x_train, y_train = np.array(x_train), np.array(y_train)

# Przekształcanie

x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

## 5.3. Importowanie potrzebnych bibliotek

```python
# pobranie potrzebnych bibliotek do tej części projektu, reszta podstawowych bibliotek została wczytana w pierwczej części

from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
```

## 5.4. Sieci neuronowe

```python

regressor = Sequential()
```

```python
# Dodanie pierwszej warstwy LSTM i regularyzacja Dropout

regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (x_train.shape[1], 1)))
regressor.add(Dropout(0.2))

# Dodanie drugiej warstwy LSTM i pewnej regularyzacji Dropout

regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))

# Dodanie trzeciej warstwy LSTM i pewnej regularyzacji Dropout

regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))

# Dodanie czwartej warstwy LSTM i pewnej regularyzacji Dropout

regressor.add(LSTM(units = 50))
regressor.add(Dropout(0.2))

# Dodawanie warstwy wyjściowej

regressor.add(Dense(units = 1))
```

Do trenowania sieci użyty zostanie optymalizator Adam, który wykorzystuje zmieniający się learning rate, zmniejszany wraz ze wzrostem liczby iteracji uczenia modelu (epok). Jest on zalecany jako optymalizator domyślny dla większości aplikacji ze względu na szybszy czas obliczeń oraz fakt, że wymaga mniejszej liczby parametrów do strojenia.

--------------------------------------------------------------------------------------------------------------------------------

###### https://keras.io/api/optimizers/adam/

```python
# Kompilowanie RNN

regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

# Dopasowanie RNN do zestawu treningowego

regressor.fit(x_train, y_train, epochs = 100, batch_size = 64)
```

```python

tsla_total = pd.concat((tsla_lstm['Close'], tsla_test['Close']), axis = 0)
inputs = tsla_total[len(tsla_total) - len(tsla_test) - 60:].values
inputs = inputs.reshape(-1,1)
inputs = sc.transform(inputs)

x_test = []

for i in range (60, 312):
    x_test.append(inputs[i-60:i, 0])
x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
predicted_stock_price = regressor.predict(x_test)
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```
